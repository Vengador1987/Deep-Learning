```python
import numpy as np
import matplotlib.pyplot as plt

import keras

# Dataset
from keras.datasets import mnist

from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.utils import to_categorical

from sklearn.metrics import confusion_matrix
```

```python
# Descargamos el dataset
# MNIST es un dataset de imagenes de numeros en escala de grises

(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train[0]
```

```python
plt.imshow(X_train[0], cmap = "gray")
plt.show()
```

```python
y_train[0]
```

```python
# Train
X_train.shape, y_train.shape
```

```python
# Test
X_test.shape, y_test.shape
```

```python
# Para adaptar cada imagen a la red neuronal debemos transformalo en un vector, es decir, un elemento de 1 columna.

# reshape
num_pixels = X_train.shape[1] * X_train.shape[2] 

# Train
X_train = X_train.reshape(X_train.shape[0], num_pixels).astype("float32") # reshape training images

# Test
X_test = X_test.reshape(X_test.shape[0], num_pixels).astype("float32") # reshape test images
```

### Normalizacion

```python
# Como tratamos con imagenes y el valor maximo es 255
# Dividimos todos los valores por 255

X_train = X_train / 255
X_test = X_test / 255
```

### One Hot Encoding



```python
# One Hot Encoding 

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
y_train[0]
```

```python
num_classes = len(y_train[0])
num_classes
```

### Neural Network

```python
# Inicializamos el modelo
model = Sequential()

# Capa de Entrada
model.add(Dense(num_pixels, activation = "relu", input_shape = (num_pixels, )))

# Capas Ocultas
model.add(Dense(100, activation = "relu"))

# Capa de Salida (Si es un problema de clasificación multiple siempre debe terminar con "softmax")
model.add(Dense(num_classes, activation = "softmax"))

# Compilamos el modelo
model.compile(optimizer = "adam", loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
# Summary
model.summary()
```

```python
# Entrenamos el modelo
model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 5, verbose = 1)

# Metricas
scores = model.evaluate(X_test, y_test, verbose = 1)
```

```python
scores
```

```python
y_hat = model.predict(X_test)
y_hat
```

```python
y_hat[0]
```

```python
np.argmax(y_hat[0])
```

```python
y_hat = [np.argmax(i) for i in y_hat]
y_hat
```

```python
y_test = [np.argmax(i) for i in y_test]
y_test
```

```python
confusion_matrix(y_test, y_hat)
```

```python
print(f"Accuracy: {scores[1]}\nError: {1 - scores[1]}")
```

```python
# Guardar Modelo
model.save("kmnist_modelo.h5")
```
